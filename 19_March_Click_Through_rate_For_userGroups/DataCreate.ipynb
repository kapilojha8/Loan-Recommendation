```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("Dataset/Recommendation Loans.csv")
```

```python
df.head()
```

```python
# Define the columns for the new dataframe
columns = ['User ID'] + [f'Loan-{i}' for i in range(1, 39)]

# Create an empty dataframe with the defined columns
new_df = pd.DataFrame(columns=columns)

# Iterate over each row in the initial dataframe
for index, row in df.iterrows():
    # Create a new row with NaN values
    new_row = pd.DataFrame([[row['UserId']] + [np.nan] * 38], columns=columns)
     
    # Fill in the loan values from the recommendations list
    for i, loan in enumerate(row['Recommendations'][1:-1].split(","), start=1):
        new_row[f'Loan-{int(loan)}'] = i
    
    # Append the new row to the new dataframe
    new_df = pd.concat([new_df, new_row], ignore_index=True)
# Sort the columns to have Loan-1, Loan-2, ... in order
new_df = new_df.reindex(sorted(new_df.columns), axis=1)
```

```python
new_df.to_csv("User Loan Recommendation matrix.csv",index=False)
```
